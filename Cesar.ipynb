```python
#Q1
states = []
shows = []
viewers = []

file_path = 'film_results.txt'

with open(file_path, 'r') as file:
    for line in file:
        
        state, show, viewer = line.strip().split(',')
        states.append(state)
        shows.append(show)
        viewers.append(viewer)

data = [[state, show, int(viewer)] for state, show, viewer in zip(states, shows, viewers)]


data_array = np.array(data, dtype=object)

with open("show_results.txt", "w") as results:
    for row in data_array:
        row_str = ','.join(map(str,row))
        results.write(row_str + "\n")
        print(row_str)

```

```python
#Q2
sorted_data = sorted(data_array.tolist(), key=lambda x: (x[0], x[1], x[2]))


states = sorted(set([item[0] for item in sorted_data]))
shows = sorted(set([item[1] for item in sorted_data]))
viewers = [item[2] for item in sorted_data]  


print("States (sorted, no duplicates):", states)
print("Shows (sorted, no duplicates):", shows)
print("Viewers (unsorted, duplicates allowed):", viewers)
```

```python
#Q3
states_array = np.array(states)
shows_array = np.array(shows)
viewers_array = np.array(viewers)

print("States Array:", states_array)
print("Shows Array:", shows_array)
print("Viewers Array:", viewers_array)
```

```python
#Q4


states_sorted = np.sort(states)
shows_sorted = np.sort(shows)

viewers_int = [int(viewer) for viewer in viewers]


total_viewers = sum(viewers_int)

print("Sorted States Array:", states_sorted)
print("Sorted Shows Array:", shows_sorted)
print("Viewers Array (as ints):", viewers_int)
print("Total Viewers:", total_viewers)
```

```python
#Q5
df = pd.DataFrame(data, columns=['State', 'Show', 'Viewers'])
show_raw_stats = pd.DataFrame(0, index=shows_sorted, columns=states_sorted)
agg_columns = ['Max', 'Min', 'Totals', 'Percent']
show_agg_stats = pd.DataFrame(index=shows_sorted, columns=agg_columns).fillna(0)

# Populate show_raw_stats with actual viewer counts
for index, row in df.iterrows():
    show = row['Show']
    state = row['State']
    viewers = row['Viewers']
    show_raw_stats.at[show, state] += viewers

for show in shows_sorted:
    show_viewers = show_raw_stats.loc[show]
    show_agg_stats.at[show, 'Max'] = show_viewers.max()
    show_agg_stats.at[show, 'Min'] = show_viewers.min()
    show_agg_stats.at[show, 'Totals'] = show_viewers.sum()
   
    total_viewership = df['Viewers'].sum()
    show_agg_stats.at[show, 'Percent'] = (show_viewers.sum() / total_viewership) * 100

print("Show Raw Stats:")
print(show_raw_stats)
print("\nShow Aggregate Stats:")
print(show_agg_stats)
```
